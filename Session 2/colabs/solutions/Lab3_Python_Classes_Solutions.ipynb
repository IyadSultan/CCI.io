## Section 5: Functions (Classes)

````python
# === SECTION 5: FUNCTIONS (CLASSES) ===

class Patient:
    """Base class representing a patient with demographics and lab values."""

    def __init__(self, name, mrn, age, hemoglobin, wbc, medications=None):
        self.name = name
        self.mrn = mrn
        self.age = age
        self.hemoglobin = hemoglobin
        self.wbc = wbc
        self.medications = medications if medications else []

    def is_anemic(self):
        """Returns True if hemoglobin is below the low threshold."""
        return self.hemoglobin < HGB_LOW

    def has_abnormal_wbc(self):
        """Returns True if WBC is outside the normal range."""
        return self.wbc > WBC_HIGH or self.wbc < WBC_LOW

    def get_alerts(self):
        """Returns a list of alert strings for abnormal values."""
        alerts = []
        if self.is_anemic():
            alerts.append(f"Low hemoglobin ({self.hemoglobin})")
        if self.has_abnormal_wbc():
            alerts.append(f"Abnormal WBC ({self.wbc})")
        return alerts

    def summary(self):
        """Returns a formatted string with all patient info."""
        return f"Patient: {self.name} ({self.mrn}) | Age: {self.age}"

    def __str__(self):
        return self.summary()
````

````python
class OncologyPatient(Patient):
    """Oncology-specific patient with tumor and treatment information."""

    def __init__(self, name, mrn, age, hemoglobin, wbc, medications=None,
                 tumor_site="Unknown", stage="Unknown", treatment_protocol="None"):
        super().__init__(name, mrn, age, hemoglobin, wbc, medications)
        self.tumor_site = tumor_site
        self.stage = stage
        self.treatment_protocol = treatment_protocol

    def is_neutropenic(self):
        """Returns True if WBC < 1.5 (chemo-related neutropenia risk)."""
        return self.wbc < NEUTROPENIC_THRESHOLD

    def get_alerts(self):
        """Override parent to include neutropenia check."""
        alerts = super().get_alerts()
        if self.is_neutropenic():
            alerts.append(f"Neutropenic (WBC {self.wbc})")
        return alerts

    def summary(self):
        """Override parent to include oncology-specific info."""
        return f"Patient: {self.name} ({self.mrn}) | Age: {self.age} | Tumor: {self.tumor_site}, Stage {self.stage}"

    def __str__(self):
        return self.summary()
````

## Section 6: Main Code

````python
# === SECTION 6: MAIN CODE ===

p1 = OncologyPatient("Ahmad", "P-1001", 58, 9.1, 1.2,
                     medications=["Cisplatin"],
                     tumor_site="Lung", stage="III",
                     treatment_protocol="Cisplatin")

p2 = OncologyPatient("Sara", "P-1002", 34, 13.5, 7.1,
                     medications=["Tamoxifen"],
                     tumor_site="Breast", stage="I",
                     treatment_protocol="Tamoxifen")

p3 = OncologyPatient("Khaled", "P-1003", 67, 7.8, 3.5,
                     medications=["Cytarabine"],
                     tumor_site="AML", stage="N/A",
                     treatment_protocol="Cytarabine")

p4 = OncologyPatient("Lina", "P-1004", 45, 11.2, 8.5,
                     medications=["Imatinib"],
                     tumor_site="CML", stage="N/A",
                     treatment_protocol="Imatinib")

p5 = OncologyPatient("Omar", "P-1005", 52, 8.5, 4.8,
                     medications=["Pembrolizumab"],
                     tumor_site="Lung", stage="II",
                     treatment_protocol="Pembrolizumab")

oncology_patients = [p1, p2, p3, p4, p5]

for p in oncology_patients:
    print(p.summary())
    alerts = p.get_alerts()
    if alerts:
        print(f"  Alerts: {', '.join(alerts)}")
    else:
        print("  Alerts: None")
````
