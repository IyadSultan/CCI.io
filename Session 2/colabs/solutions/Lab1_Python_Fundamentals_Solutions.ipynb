### Clinical Scenario
> You are a data analyst at KHCC and receive a daily list of patients with their hemoglobin and WBC counts. You need to flag abnormal values automatically instead of checking them manually.

### Reference Thresholds
- Hemoglobin < 10.0 g/dL = **LOW**
- WBC > 11.0 x10^9/L = **HIGH**
- WBC < 4.0 x10^9/L = **LOW**

---
## Cell 1 - Variables

````python
# === CELL 1: VARIABLES ===

patient_name = "Ahmad"
mrn = "P-1001"
hemoglobin = 9.1
wbc = 6.2
on_chemo = True

print("Patient Name:", patient_name)
print("MRN:", mrn)
print("Hemoglobin:", hemoglobin)
print("WBC:", wbc)
print("On Chemo:", on_chemo)
````

**Expected output:**
```
Patient Name: Ahmad
MRN: P-1001
Hemoglobin: 9.1
WBC: 6.2
On Chemo: True
```

---
## Cell 2 - Dictionary

````python
# === CELL 2: DICTIONARY ===

patient = {
    "name": "Ahmad",
    "mrn": "P-1001",
    "hgb": 9.1,
    "wbc": 6.2,
    "on_chemo": True
}

print("Name:", patient["name"])
print("MRN:", patient["mrn"])
print("Hemoglobin:", patient["hgb"])
print("WBC:", patient["wbc"])
print("On Chemo:", patient["on_chemo"])
print("\nFull record:", patient)
````

---
## Cell 3 - List of Patients

````python
# === CELL 3: LIST OF PATIENTS ===

patients = [
    {"name": "Ahmad",  "mrn": "P-1001", "hgb": 9.1,  "wbc": 6.2,  "on_chemo": True},
    {"name": "Sara",   "mrn": "P-1002", "hgb": 13.5, "wbc": 12.3, "on_chemo": False},
    {"name": "Khaled", "mrn": "P-1003", "hgb": 7.8,  "wbc": 3.5,  "on_chemo": True},
    {"name": "Lina",   "mrn": "P-1004", "hgb": 11.2, "wbc": 7.1,  "on_chemo": False},
    {"name": "Omar",   "mrn": "P-1005", "hgb": 8.5,  "wbc": 9.8,  "on_chemo": True},
]

print(f"Total patients: {len(patients)}")
````

---
## Cell 4 - Alert Loop

````python
# === CELL 4: ALERT LOOP ===

for p in patients:
    if p["hgb"] < 10.0:
        print(f"ALERT: Patient {p['name']} - Hemoglobin {p['hgb']} (LOW)")
    if p["wbc"] > 11.0:
        print(f"ALERT: Patient {p['name']} - WBC {p['wbc']} (HIGH)")
    if p["wbc"] < 4.0:
        print(f"ALERT: Patient {p['name']} - WBC {p['wbc']} (LOW)")
````

**Expected output:**
```
ALERT: Patient Ahmad - Hemoglobin 9.1 (LOW)
ALERT: Patient Sara - WBC 12.3 (HIGH)
ALERT: Patient Khaled - Hemoglobin 7.8 (LOW)
ALERT: Patient Khaled - WBC 3.5 (LOW)
ALERT: Patient Omar - Hemoglobin 8.5 (LOW)
```

---
## Cell 5 - Summary

````python
# === CELL 5: SUMMARY ===

abnormal_count = 0

for p in patients:
    if p["hgb"] < 10.0 or p["wbc"] > 11.0 or p["wbc"] < 4.0:
        abnormal_count += 1

print(f"Summary: {abnormal_count} of {len(patients)} patients have abnormal values")
````

**Expected output:**
```
Summary: 4 of 5 patients have abnormal values
```

---
## Stretch Challenge - Severity Classification

````python
# === STRETCH CHALLENGE ===

for p in patients:
    if p["hgb"] < 7.0:
        severity = "CRITICAL"
    elif p["hgb"] < 10.0:
        severity = "LOW"
    else:
        severity = "NORMAL"
    print(f"Patient {p['name']}: Hemoglobin {p['hgb']} - {severity}")
````

**Expected output:**
```
Patient Ahmad: Hemoglobin 9.1 - LOW
Patient Sara: Hemoglobin 13.5 - NORMAL
Patient Khaled: Hemoglobin 7.8 - LOW
Patient Lina: Hemoglobin 11.2 - NORMAL
Patient Omar: Hemoglobin 8.5 - LOW
```
