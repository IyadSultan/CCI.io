## Section 6: Main Code

### Step 1: Create the Demographics DataFrame

````python
# Step 1: Demographics
demographics = pd.DataFrame([
    {"mrn": "P-1001", "name": "Ahmad",  "age": 58, "gender": "M", "diagnosis": "Lung Cancer"},
    {"mrn": "P-1002", "name": "Sara",   "age": 34, "gender": "F", "diagnosis": "Breast Cancer"},
    {"mrn": "P-1003", "name": "Khaled", "age": 67, "gender": "M", "diagnosis": "AML"},
    {"mrn": "P-1004", "name": "Lina",   "age": 45, "gender": "F", "diagnosis": "CML"},
    {"mrn": "P-1005", "name": "Omar",   "age": 52, "gender": "M", "diagnosis": "Lung Cancer"},
    {"mrn": "P-1006", "name": "Rania",  "age": 41, "gender": "F", "diagnosis": "Breast Cancer"},
])

print("Demographics:")
print(demographics)
print(f"\nShape: {demographics.shape}")
````

### Step 2: Create the Labs DataFrame

````python
# Step 2: Labs (note: P-1006 has no labs yet)
labs = pd.DataFrame([
    {"mrn": "P-1001", "hemoglobin": 9.1,  "wbc": 6.2, "platelets": 180},
    {"mrn": "P-1002", "hemoglobin": 13.5, "wbc": 7.1, "platelets": 220},
    {"mrn": "P-1003", "hemoglobin": 7.8,  "wbc": 1.2, "platelets": 45},
    {"mrn": "P-1004", "hemoglobin": 11.2, "wbc": 8.5, "platelets": 195},
    {"mrn": "P-1005", "hemoglobin": 8.5,  "wbc": 4.8, "platelets": 160},
])

print("Labs:")
print(labs)
````

### Step 3: Create the Medications DataFrame

````python
# Step 3: Medications
medications = pd.DataFrame([
    {"mrn": "P-1001", "drug": "Cisplatin",      "on_chemo": True},
    {"mrn": "P-1002", "drug": "Tamoxifen",      "on_chemo": False},
    {"mrn": "P-1003", "drug": "Cytarabine",     "on_chemo": True},
    {"mrn": "P-1004", "drug": "Imatinib",       "on_chemo": False},
    {"mrn": "P-1005", "drug": "Pembrolizumab",  "on_chemo": True},
    {"mrn": "P-1006", "drug": "Letrozole",      "on_chemo": False},
])

print("Medications:")
print(medications)
````

### Step 4: Merge all three DataFrames

````python
# Step 4: Merge
df = pd.merge(demographics, labs, on="mrn", how="left")
df = pd.merge(df, medications, on="mrn", how="left")

print("Merged DataFrame:")
print(df)
print(f"\nMerged shape: {df.shape}")
````

### Step 5: Check for missing data

````python
# Step 5: Check missing values
print("Missing values:")
print(df.isna().sum())
````

**Expected:** hemoglobin 1, wbc 1, platelets 1 (Rania has no labs)

### Step 6: Filter critical patients

````python
# Step 6: Filter critical patients
critical = df[(df["hemoglobin"] < HGB_LOW) & (df["on_chemo"] == True)]

print("\nCritical patients (low HGB + on chemo):")
for _, row in critical.iterrows():
    print(f"  {row['mrn']} {row['name']} -- HGB: {row['hemoglobin']}, Drug: {row['drug']}")
````

**Expected output:**
```
Critical patients (low HGB + on chemo):
  P-1001 Ahmad -- HGB: 9.1, Drug: Cisplatin
  P-1003 Khaled -- HGB: 7.8, Drug: Cytarabine
  P-1005 Omar -- HGB: 8.5, Drug: Pembrolizumab
```

### Step 7: Group by diagnosis

````python
# Step 7: Group by diagnosis
print("\nMean hemoglobin by diagnosis:")
mean_hgb = df.groupby("diagnosis")["hemoglobin"].mean()
print(mean_hgb)
````

**Expected output:**
```
Mean hemoglobin by diagnosis:
diagnosis
AML            7.80
Breast Cancer  13.50
CML            11.20
Lung Cancer     8.80
```
