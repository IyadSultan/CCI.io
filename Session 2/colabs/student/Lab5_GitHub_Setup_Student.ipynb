## Step 1: Create Your GitHub Account

If you have not already created a GitHub account:

1. Open a new browser tab and go to [github.com](https://github.com)
2. Click **Sign up**
3. Use your KHCC or personal email
4. Choose a **professional username** (e.g., `ahmad-khcc`, not `xXcoder420Xx`)
5. Complete the account setup

**Already have an account?** Skip to Step 2.

---
## Step 2: Create a New Repository

1. Click the **+** icon in the top-right corner of GitHub
2. Select **New repository**
3. Fill in the details:
   - **Name:** `cci-course-notebooks`
   - **Description:** "CCI Prompt Engineering & Clinical AI Development -- Course Notebooks"
   - **Visibility:** Public (for course purposes) or Private
   - Check: **Add a README file**
   - Add .gitignore: Select **Python**
4. Click **Create repository**

### Why .gitignore matters
The Python .gitignore template automatically excludes:
- `__pycache__/` directories
- `.env` files (API keys)
- Data files that should not be uploaded

**IMPORTANT:** Never push notebooks containing real patient data (MRNs, names, PHI) to GitHub.

---
## Step 3: Push Your First Notebook from Colab

1. Open your **Session2_Lab2_Notebook_Template.ipynb** in Colab (or any completed lab notebook)
2. Go to **File > Save a copy in GitHub**
3. If prompted, authorize Google Colab to access your GitHub account
4. Select your `cci-course-notebooks` repository
5. Set the file path: `session_2/Lab2_Notebook_Template.ipynb`
6. Write a descriptive commit message:
   ```
   Add Session 2 Lab 2 -- Clinical notebook template with 9-section structure
   ```
7. Click **OK**

---
## Step 4: Verify on GitHub

1. Go to your repository on github.com
2. Navigate to the `session_2/` folder
3. Click on the notebook -- GitHub renders .ipynb files
4. Check the commit history (clock icon or "Commits" link)
5. You should see your commit message

---
## Step 5: Push a Second Notebook

Repeat Step 3 with a different notebook (e.g., Lab 1 or Lab 3).
Use a **different commit message** to see multiple commits in history.

Example commit messages:
- `Add Session 2 Lab 1 -- Python fundamentals with patient data variables and alert loop`
- `Add Session 2 Lab 3 -- Patient and OncologyPatient class hierarchy`
